## Coding Session I - 30 minutes

From the Fitbit data, we want to visually understand: 

1. **How the average number of steps change by month. Use the appropriate visualization to show the median steps by month?**
1. **Is Fitbitter more active on weekend or workdays?**

##### All plots must be in the same jupyter notebook cell.

```python
# Driver 1 enter codes here
# Aquí saco la media total solo por curiosidad

Steps_ave = fitbit['Steps'].mean()
Steps_ave
```

```python
# Agrupo por mes y hago la media de todas las columnas

grouped_months = fitbit.groupby('Months_encoded')
mean_by_months = grouped_months.mean()
mean_by_months
```

```python
# Saco solo la columna de 'Steps'

mean_steps = grouped_months['Steps'].mean()
mean_steps
```

```python
f, ax = plt.subplots()

x = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
l = mean_steps.index
y = mean_steps.values

ax.bar(l,y)
ax.set_xticks(l)
ax.set_xticklabels(x, rotation='vertical')

plt.show()
```

```python
# Hacemos la misma operación pero agrupando por días

grouped_weekdays = fitbit.groupby('Days_encoded')
mean_by_weekdays = grouped_weekdays.mean()
mean_by_weekdays
```

```python
mean_weekdays = grouped_weekdays['Steps'].mean()
mean_weekdays
```

```python
f, ax = plt.subplots()

x = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday','Saturday','Sunday']
l = mean_weekdays.index
y = mean_weekdays.values

ax.bar(l,y)
ax.set_xticks(l)
ax.set_xticklabels(x, rotation='vertical')

plt.show()
```

Como conclusión sacamos que el sábado es el día que más pasos se dan en toda la semana, y el domingo, efectivamente, es el día que menos

```python
cols=['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary','Steps']
df=fitbit[cols]
fig,axes = plt.subplots(figsize = (16,8), nrows = 1, ncols = 3)
for i in range(3):
    plt.sca(axes[i])
    plt.scatter(df.iloc[:,i],df['Steps'], alpha = 0.8)
    plt.xlabel(df.iloc[:,i].name)
    plt.ylabel('Steps')
```
