# Challenge 1

#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Primero ploteo por separado
x = np.arange(0,100)
y = x ** 2
plt.plot(x, y, color="b", ls='--')
```

```python
x = np.arange(0,100)
z = x**2
plt.plot(x, z, color="b")
```

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)

plt.subplot(2,1,1)
plt.plot(x,y,lw=3,color="b", ls='--')
plt.subplot(2,1,2)
plt.plot(x,z,lw=3, color="r",ls='-')
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)

fig,axes=plt.subplots(1,2)
axes[0].plot(x,y,color="b", lw=3, ls='--')
axes[1].plot(x,z,color="r", lw=3, ls='-')
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here

fig,axes=plt.subplots(1,2, figsize=(15,10))
axes[0].plot(x,y,color="b", lw=3, ls='--')
axes[1].plot(x,z,color="r", lw=3, ls='-')
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here

fig,axes=plt.subplots(1,2, figsize=(15,10))
axes[0].plot(x,y,color="b", lw=3, ls='--')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="r", lw=3, ls='-')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```

#### Plot both `y=x^2` and `y=exp(x)`in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here

fig,axes=plt.subplots(1,2, figsize=(15,10))
axes[0].plot(x,x**2, np.exp(x))

axes[1].plot(x,x**2, np.exp(x))
axes[1].set_yscale("log")
```
