```python
# Leemos el archivo 'CSV' con ayuda de un 'encoding'
# '../input' para indicar la nueva ruta del 'CSV'

shark_df = pd.read_csv('../input/GSAF5.csv', encoding = 'iso8859_15')
```

```python
null_cols = shark_df.isnull().sum()
null_cols
```

```python
#Comprobamos la longitud de nuestro dataset:
len(shark_df)
```

```python
#Ahora sacamos una lista con las columnas que tienen mÃ¡s de 1000 nulos:

drop_cols = list(null_cols[null_cols > 1000].index)
drop_cols
```

```python
#Una vez sabemos cuales son las columnas que queremos eliminar, procedemos a ello:
shark_df = shark_df.drop(drop_cols, axis=1)
```
